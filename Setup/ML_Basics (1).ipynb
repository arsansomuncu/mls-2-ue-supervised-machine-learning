```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn import datasets

from pyMLaux import plot_2d_prediction
```

## Linear Regression: Concentration (Mockup Data)

```python
xVal = np.array(range(1, 6), dtype='float')
yVal = 50 + 12 * (xVal - 3) + np.random.normal(0, 4, len(xVal))
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(xVal, yVal)
ax.set_xlim(0, 6)
ax.set_ylim(0, 100)
ax.set_xlabel('Concentration (%)')
ax.set_ylabel('Absorption (%)')
plt.show()
```

```python
lin_model = LinearRegression()
lin_model.fit(xVal.reshape(len(xVal), 1), yVal)
```

```python
xTest = np.array([0, 6])
yTest = lin_model.predict(xTest.reshape(2, 1))

fig = plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.plot(xTest, yTest, linewidth=2, linestyle='dashed', color='darkred')
ax.scatter(xVal, yVal)
ax.set_xlim(0, 6)
ax.set_ylim(0, 100)
ax.set_xlabel('Concentration (%)')
ax.set_ylabel('Absorption (%)')
plt.show()
```

## Exploring Data: Fish Example

```python
data_dir = ""
fish_data_raw = pd.read_csv('fish.csv')
fish_data = {'data': np.array(fish_data_raw.iloc[:, :-1]),
             'target': (np.array(fish_data_raw.iloc[:, -1]) + 1) / 2,
             'feature_names': fish_data_raw.columns[:-1],
             'target_names': ['Salmon', 'Sea bass']}
```

```python
for j, f in enumerate(fish_data['feature_names']):
    for i, cl in enumerate(fish_data['target_names']):
        plt.hist(fish_data['data'][:, j][fish_data['target'] == i], bins=10,
                 alpha=0.5, range=(0, 1), label=cl)
    plt.title(f)
    plt.legend(loc='upper right')
    plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(fish_data['data'][:, 0], fish_data['data'][:, 1], c=fish_data['target'], cmap='bwr')
plt.show()
```

```python
svm_model1 = SVC(kernel='linear', C=100)
svm_model1.fit(X=fish_data['data'], y=fish_data['target'])
```

```python
plot_2d_prediction(fish_data['data'], fish_data['target'], svm_model1.decision_function,
                   xval=np.linspace(0, 1, 101), yval=np.linspace(0, 1, 101))
```

```python
svm_model2 = SVC(kernel='rbf', C=12, gamma=1.8)
svm_model2.fit(X=fish_data['data'], y=fish_data['target'])
```

```python
plot_2d_prediction(fish_data['data'], fish_data['target'], svm_model2.decision_function,
                   xval=np.linspace(0, 1, 101), yval=np.linspace(0, 1, 101))
```

```python
svm_model3 = SVC(kernel='rbf', C=1000000, gamma=20)
svm_model3.fit(X=fish_data['data'], y=fish_data['target'])
```

```python
plot_2d_prediction(fish_data['data'], fish_data['target'], svm_model3.decision_function,
                   xval=np.linspace(0, 1, 101), yval=np.linspace(0, 1, 101))
```
